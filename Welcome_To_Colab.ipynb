```
import pandas as pd
import matplotlib.pyplot as plt

# Sample data in dictionary format
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Page Views': [828, 7065, 5861, 7163, 9432],
    'Unique Visitors': [1261, 4225, 3286, 651, 548],
    'Bounce Rate': [54.4200085, 31.58388663, 68.28470317, 60.20317463, 37.96324748]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Page Views per Visitor
df['Page Views per Visitor'] = df['Page Views'] / df['Unique Visitors']

# Plot Page Views and Unique Visitors over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Page Views'], label='Page Views', marker='o')
plt.plot(df['Date'], df['Unique Visitors'], label='Unique Visitors', marker='o')
plt.title('Page Views and Unique Visitors Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Plot Bounce Rate Over Time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Bounce Rate'], label='Bounce Rate (%)', marker='o', color='red')
plt.title('Bounce Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Bounce Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Show DataFrame for reference
print(df)
```
